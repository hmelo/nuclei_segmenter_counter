# Cell segmentation and counting DEMO

In this demo we show an end-to-end deep learning pipeline for segmentation and counting of digital pathology images.

## 1. Load images

```python
from celllib import *
# load and visualize image

TEST_PATH = '../Nuclei/stage1_test/'
test_ids = next(os.walk(TEST_PATH))[1]
TEST_DATA,test_img_sizes = read_test_data(TEST_PATH,test_ids)
```

## 2. Perform Segmentation

```python
# run prediction segmentation

model = get_unet()
model.load_weights('unet_simple_model_150epochs_weights.h5')
preds_test = model.predict(TEST_DATA, verbose=1)
```

```python
# visualize segmentation
img=12
show_segment(TEST_DATA,preds_test, img)
```

## 3. Perform Cell Counting

```python
## perform counting
count_cells(preds_test[img], t=65)
```
